#### 6. Cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function

###### **To Cleanse year**

```python
def cleanse_year(year_str):
    if "*" in year_str:
        year = year_str.replace("*", "")
    else:
        year = year_str

    return int(year)
```

```python
df["year_cleansed"] = df["Census Year"].apply(cleanse_year)

df
```

###### **To Cleanse population**

```python
def cleanse_pop(pop_str):
    if "," in pop_str:
        pop = pop_str.replace(",", "")
    else:
        pop = pop_str

    return int(pop)
```

```python
df["population_cleansed"] = df["Total Population"].apply(cleanse_pop)

df
```

###### **To Cleanse cost**

```python
def cleanse_cost(cost_str):
    if "$" in cost_str:
        cost_str = cost_str.replace("$", "")
    if "," in cost_str:
        cost_str = cost_str.replace(",", "")
    if "Billion" in cost_str:
        cost_str = cost_str.replace(" Billion", "")
        cost_str = float(cost_str) * 1000000000

    return int(cost_str)
```

```python
df["cost_cleansed"] = df["Census Cost"].apply(cleanse_cost)

df
```

###### **To Cleanse average cost per person**

```python
def cleanse_average(average_str):
    if "$" in average_str:
        average_str = average_str.replace("$", "")
    if "," in average_str:
        average_str = average_str.replace(",", "")
    if "cents" in average_str:
        average_str = average_str.replace(" cents", "")
        average_str = float(average_str) / 100                    # to convert in dollars

    return float(average_str)
```

```python
df["average_cost_cleansed"] = df["Average Cost Per Person"].apply(cleanse_average)

df
```
